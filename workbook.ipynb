```python
# Data
min_place = [10, 25, 50, 100, 250, 500, 1000, 2000, 3000, 5000]
coeff = [1.61, 1.52, 1.43, 1.32, 1.16, 1.0, 0.81, 0.6, 0.43, 0.31]
```

```python
plt.figure(figsize=(10, 6))
plt.plot(min_place, coeff, marker='o', linestyle='-', color='b')

# Add titles and labels
plt.title('Line Graph of Given Pairs of Numbers')
plt.xlabel('X values')
plt.ylabel('Y values')

# Show the plot
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Define N and Q range
N = 36
Q = np.arange(0, 37)

# Create the plot
plt.figure(figsize=(20, 8))

# Iterate over each pair of (min_place, coeff) and plot the step line
for min_p, c in zip(min_place, coeff):
    trueDL = (1 - np.minimum(Q / c, N) / N) * 10
    answered_ratio = Q / N
    plt.step(answered_ratio, (10-trueDL), label=f'min_place={min_p}, coeff={c}', where='mid')

# Add titles and labels
plt.title('Step Graph of trueDL')
plt.xlabel('answered_ratio (Q / N)')
plt.ylabel('trueDL')

# Add legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
